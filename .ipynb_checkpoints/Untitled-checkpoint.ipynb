```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
city_import = "Resources/cities.csv" #import csv file
citydata = pd.read_csv(city_import) #read csv file
citydata.head()
```

```python
#Creating a Scatterplot using Seaborn
sns.relplot(x="Lat", y="Max Temp", 
            data=citydata, alpha = 0.7, edgecolor = "Black", 
            linewidths = None)
plt.title("City Latitude vs. Max Temperature", fontsize=16)
plt.xlabel("Latitude", fontsize=16)
plt.ylabel("Max Temperature(F)", fontsize=16)
plt.savefig('Resources/assets/images/fig1.png')
plt.show()
```

```python
#Creating a  Scatterplot using Seaborn
sns.relplot(x="Lat", y="Humidity", 
            data=citydata, alpha = 0.7, edgecolor = "Black",  
            linewidths = None)
sns.set_style("whitegrid")
plt.title("City Latitude vs. Humidity", fontsize=16)
plt.ylabel("Humidity(%)", fontsize=16)
plt.xlabel("Latitude", fontsize=16)
plt.savefig('Resources/assets/images/fig2.png')
plt.show()
```

```python
#Creating a  Scatterplot using Seaborn
sns.relplot(x="Lat", y="Cloudiness", 
            data=citydata, alpha = 0.7, edgecolor = "Black",  
            linewidths = None)
sns.set_style("whitegrid")
plt.title("City Latitude vs. Cloudiness", fontsize=16)
plt.ylabel("Cloudiness(%)", fontsize=16)
plt.xlabel("Latitude", fontsize=16)
plt.savefig('Resources/assets/images/fig3.png')
plt.show()
```

```python
#Creating a  Scatterplot using Seaborn
sns.relplot(x="Lat", y="Wind Speed", 
            data=citydata, alpha = 0.7, edgecolor = "Black",  
            linewidths = None)
sns.set_style("whitegrid")
plt.title("City Latitude vs. Wind Speed", fontsize=16)
plt.ylabel("Wind Speed(mph)", fontsize=16 )
plt.xlabel("Latitude", fontsize=16)
plt.savefig('Resources/assets/images/fig4.png')
plt.show()
```
